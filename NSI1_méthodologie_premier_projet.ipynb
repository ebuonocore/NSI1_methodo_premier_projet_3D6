## ‚öÑ Probl√®me 5 : Jeu √† deux joueurs

Notre programme commence √† √™tre correct.  
Il nous reste √† g√©rer le second joueur et √† comparer les r√©sultats pour d√©signer le vainqueur.  
Nous pourrions copier/coller le programme et juste le modifier pour identifier deux variables diff√©rentes <code>nb_tentatives1</code> et <code>nb_tentatives2</code> pour chacun des joueurs, mais nous allons factoriser un peu ce code en d√©finissant une fonction.

La fonction <code>partie()</code> prend l'entier <code>numero_joueur</code> en param√®tre.  
Elle g√®re la partie du joueur et renvoie √† la fin le nombre de tentatives.

````python
from random import randrange

def partie(...):  # Pr√©cisez ici le param√®tre d'entr√©e.
    """ Prend l'entier numero_joueur en param√®tre.
        G√®re la partie du joueur et renvoie √† la fin le nombre de tentatives.
    """
    D_1 = randrange(1,7)
    D_2 = randrange(1,7)
    D_3 = randrange(1,7)
    somme = D_1 + D_2 + D_3
    texte = "R√©sultat du premier lanc√© du joueur " + str(numero_joueur) + " : "
    texte+= str(D_1) + " + " + str(D_2) + " + " + str(D_3) + " = " + str(somme)
    nb_tentatives = 1  # Initialisation de la variable

    while somme != 10:
        texte += "\nQuel d√© voulez-vous relancer ? (saisssez 1, 2 ou 3)"
        choix = input(texte)
        if choix == "1":
            D_1 = randrange(1,7)
        elif choix == "2":
            D_2 = randrange(1,7)
        elif choix == "3":
            D_3 = randrange(1,7)
        else:
            print("Mauvais choix, recommencez.")
            nb_tentatives -= 1  # Compense l'incr√©ment superflu
        nb_tentatives += 1  # Incr√©mentation du nombre de tentatives.
        somme = D_1 + D_2 + D_3
        texte = "R√©sultat du lanc√© : " 
        texte += str(D_1) + " + " + str(D_2) + " + " + str(D_3) + " = " + str(somme)
    ...  # Valeur √† renvoyer

# Corps du programme
nb_tentatives1 = partie(1)
print("Le joueur 1 fait 10 au bout de ", nb_tentatives1, " coups.")
nb_tentatives2 = partie(2)
print("Le joueur 2 fait 10 au bout de ", nb_tentatives2, " coups.")
if nb_tentatives1 > nb_tentatives2:
    print("Le joueur 1 a gagn√© avec ",nb_tentatives1, "coups, contre ", nb_tentatives2)
elif nb_tentatives1 < nb_tentatives2:
    print("Le joueur 2 a gagn√© avec ",nb_tentatives2, "coups, contre ", nb_tentatives1)
else:
    print("Egalit√© : ", nb_tentatives2, "coups chacun.")
````
